```python
from bayes_regression import Regressor
```

```python
import datetime
from datetime import date
```

```python
from model import (
    init,
    Brand,
    MarketType,
    IndustoryType,
    DayChart,
)
from model import database_proxy as db
```

```python
init(
    MODE = 'MYSQL',
    DROP_DB = False,
    CREATE_DB = False
)
db.connect()
```

```python
# 情報通信の方々のみ取得
brands = Brand.select().join(IndustoryType).where(IndustoryType.id == 12)
```

```python
target = brands.where(Brand.name==u"ガンホー").get()
```

```python
print target.name
```

```python
competitors = brands.where(
    (Brand.name!=u"ガンホー") &
    (Brand.id!=3912) )[0:]
```

```python
len(competitors)
```

```python
s_date = date(2013,1,1)
e_date = date(2013,1,15)
test_s_date = date(2013,1,16)
test_e_date = date(2013,2,1)
```

```python
t_step = 1
attr = 'max_value'
```

```python
r = Regressor(target, competitors, s_date, e_date, t_step, attr)
```

```python
r.get_data(update_self=True)
```

```python
r.train()
```

```python
r.test(test_s_date, test_e_date)
```

```python
r.plot_forecast()
```

```python
r.plot_weight()
```

```python
r.plot_log_likelihood()
```

```python
r.plot_sqerr()
```
